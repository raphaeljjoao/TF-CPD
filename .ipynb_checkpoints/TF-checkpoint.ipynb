```python
import pandas as pd
tabela = pd.read_csv("dados/MICRODADOS_ENEM_2000.csv", sep=",", encoding="utf-8")

pd.set_option('display.max_columns', None)
display(tabela)
```

```python
# LISTA OS MUNICIPIOS DISPONIVEIS NOS DADOS
def lista_municipios():
    municipios = tabela[["NO_MUNICIPIO_RESIDENCIA"]].groupby("NO_MUNICIPIO_RESIDENCIA").sum()
    pd.set_option('display.max_rows', None)
    
    return municipios
```

```python
# PESQUISANDO UM MUNICIPIO
def busca_municipio(busca):
    nome_municipio = busca
    municipio = tabela[tabela.NO_MUNICIPIO_RESIDENCIA==nome_municipio].count()[0]
    print("Número de aparições do municipio '{}' = {}".format(nome_municipio, municipio))
    tabela_municipio = tabela[tabela.NO_MUNICIPIO_RESIDENCIA==nome_municipio]

    return tabela_municipio

# CHECANDO CANDIDATOS QUE ESTIVERAM PRESENTES E AUSENTES NA PROVA
# PARA OBTER CANDITOS PRESENTES BASTA INSERIR 1 NA VARIAVEL VALOR
# PARA OBTER CANDITOS AUSENTES BASTA INSERIR 0 NA VARIAVEL VALOR
def obter_candidatos_por_status(municipio, valor):
    if(valor == 0):
        tabela_municipio = busca_municipio(municipio)
        candidatos_ausentes = tabela_municipio[tabela_municipio.TP_PRESENCA==valor]
        num_ausentes = candidatos_ausentes.count()[0]
        display(candidatos_ausentes)
        print("Número de candidatos ausentes durante a prova no municipio de '{}' = {}".format(municipio, num_ausentes))
        return candidatos_ausentes
    elif(valor == 1):
        tabela_municipio = busca_municipio(municipio)
        candidatos_presentes = tabela_municipio[tabela_municipio.TP_PRESENCA==valor]
        num_presentes = candidatos_presentes.count()[0]
        display(candidatos_presentes)
        print("Número de candidatos presentes durante a prova no municipio de '{}' = {}".format(municipio, num_presentes))
        return candidatos_presentes

obter_candidatos_por_status("JOINVILLE", 0)
```
